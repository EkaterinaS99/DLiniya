```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
lmd = 1 # длина волны для вычисления волнового числа 2pi/lmd
W = 50 # волновое сопротивление линии
Rn = W * 1
Xn = W * 0.5
Zn = Rn + Xn*1j # нормировано к W
Un=1 
```

```python
In=Un/Zn
z=np.linspace(0,8,1000)
I=In*np.cos(2*np.pi/lmd*z)+Un/W*np.sin(2*np.pi/lmd*z)
U=Un*np.cos(2*np.pi/lmd*z)+In*W*np.sin(2*np.pi/lmd*z)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(z,np.abs(I),"r")
ax.grid()
ax.set_xlim([1.5,0])
ax.set_xlabel('z, м')
ax.set_ylabel('|I|, А')
plt.title("Распределение амплитуд тока")
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(z,np.abs(U))
ax.grid()
ax.set_xlim([1.5,0])
ax.set_xlabel('z, м')
ax.set_ylabel('|U|, В')
plt.title("Распределение амплитуд напряжения")
```

```python
G=(Zn-W)/(Zn+W)
Ksv=(1+abs(G))/(1-abs(G))
Kbv=1/Ksv
print(Kbv,Ksv)
```
